```python
new_df=df[['id','host.host_name','room_type','address.country','host.host_neighbourhood','address.location.coordinates','last_review','price','availability.availability_365','review_scores.review_scores_rating']]
```

```python
new_df=new_df.rename(columns={"host.host_name": "host_name", "address.country":"country" , "host.host_neighbourhood" : "neighbourhood", "last_review":"visiting_dt", "availability.availability_365" : "availabile_365" , "review_scores.review_scores_rating" : "reviews"})
new_df
```

```python
print(new_df.dtypes)
print(new_df.shape)
```

column -> address.location.coordinates latitude & longitude separation

```python
new_df[['latitude', 'longitude']] = new_df['address.location.coordinates'].str.strip('()').str.split(', ', expand=True)
new_df[['latitude', 'longitude']] = new_df[['latitude', 'longitude']].applymap(lambda x: x.strip("[]"))
# Print the updated DataFrame
print(new_df[['latitude', 'longitude']])
```

```python
new_df.drop(columns=['address.location.coordinates'], inplace=True)
```

```python
# Convert the latitude and longitude columns to numeric (if needed)
print(new_df["latitude"].dtype)
print(new_df["latitude"].dtype)
new_df['latitude'] = pd.to_numeric(new_df['latitude'])
new_df['longitude'] = pd.to_numeric(new_df['longitude'])
```

```python
print(new_df["latitude"].dtype)
print(new_df["latitude"].dtype)
```

```python
new_df.isnull().sum()
```

```python
#reviews have 1474 missing values so we should clean it..

# Calculate the median of the 'reviews' column
median_value = new_df['reviews'].median()

# Fill NaN values in the 'reviews' column with the median value
new_df['reviews'] = new_df['reviews'].fillna(median_value)

new_df['reviews'] = pd.to_numeric(new_df['reviews'])
new_df['reviews'] = new_df['reviews'].astype('int64')
new_df['price'] = pd.to_numeric(new_df['price'])
new_df['price'] = new_df['price'].astype('int64')
```

```python
#neighbourhood data cleaning
#we don't need exact neighbourhood name, but we need that particular values of it
#so i fill it with "X" 
new_df['neighbourhood'].fillna('X', inplace=True)
```

```python
#visiting date, it is used for calculating season wise analysis

new_df['visiting_dt'] = pd.to_datetime(new_df['visiting_dt'])
new_df['date'] = new_df['visiting_dt'].dt.date
```

```python

new_df.drop(columns=['visiting_dt'], inplace=True)
```

```python
# Fill null values with the most recent non-null value (forward-fill)
new_df['date'].fillna(method='ffill', inplace=True)
```

```python
new_df.isnull().sum()
```

```python
final_df=new_df

# Sort the DataFrame based on the 'column_name' column in ascending order
final_df=final_df.sort_values(by='id', ascending=True,)
final_df
```
